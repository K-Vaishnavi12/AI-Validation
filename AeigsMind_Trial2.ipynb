```python
class Orchestrator:
    def enrich(self, provider: Dict) -> Dict:
        log("ORCHESTRATION", f"Enriching provider {provider['name']}")
        # Simulate API calls for NPI, state license, and location data
        provider['npi_data'] = {"npi_valid": random.choice([True, False]), "registered_name": ''.join(random.choices(string.ascii_uppercase, k=6))}
        provider['state_data'] = {"license_active": random.choice([True, False]), "disciplinary_action": random.choice([True, False, False])}
        provider['location_data'] = {"lat": random.uniform(-90, 90), "lng": random.uniform(-180, 180), "address_valid": random.choice([True, False])}
        return provider

class WatchdogScanner:
    def scan(self, provider: Dict) -> Dict:
        log("WATCHDOG", f"Scanning provider {provider['name']}")
        issues = []
        if not provider['npi_data']['npi_valid']:
            issues.append("Invalid NPI")
        if not provider['state_data']['license_active'] or provider['state_data']['disciplinary_action']:
            issues.append("License Issue")
        if not provider['location_data']['address_valid']:
            issues.append("Invalid Address")
        provider['validation_issues'] = issues
        return provider

class CredentialVerifier:
    def verify(self, provider: Dict) -> Dict:
        log("VERIFIER", f"Verifying provider {provider['name']}")
        score = 100
        if "License Issue" in provider['validation_issues']:
            score -= 50
        provider['credential_score'] = score
        return provider

class ConfidenceScorer:
    def score(self, provider: Dict) -> Dict:
        log("CONFIDENCE", f"Scoring provider {provider['name']}")
        confidence = provider['credential_score']
        if "Invalid NPI" in provider['validation_issues']:
            confidence -= 20
        if "Invalid Address" in provider['validation_issues']:
            confidence -= 20
        provider['confidence_score'] = max(0, confidence)
        return provider

class OutputLayer:
    def process(self, processed_data: List[Dict]):
        log("OUTPUT", "Generating reports")
        df = pd.DataFrame(processed_data)

        provider_report = df[["provider_id", "confidence_score"]].copy()
        provider_report["issues"] = df["validation_issues"].apply(lambda x: x if x else [])

        flagged_providers = df[df["confidence_score"] < 60]["provider_id"].tolist()

        return provider_report, flagged_providers

orchestrator = Orchestrator()
watchdog = WatchdogScanner()
verifier = CredentialVerifier()
scorer = ConfidenceScorer()
output = OutputLayer()

processed = []

for _, row in raw_provider_data.iterrows():
    provider = row.to_dict()
    provider = orchestrator.enrich(provider)
    provider = watchdog.scan(provider)
    provider = verifier.verify(provider)
    provider = scorer.score(provider)
    processed.append(provider)

provider_report, flagged_providers = output.process(processed)
```

```python
import gradio as gr
import pandas as pd
import random

def process_providers(file):

    df = pd.read_csv(file.name)

    df['Confidence'] = [random.randint(50, 100) for _ in range(len(df))]

    df['Flagged'] = df['Confidence'].apply(lambda x: 'Yes' if x < 75 else 'No')

    health_report = df[['Provider', 'State', 'Confidence']]

    flagged_list = df[df['Flagged']=='Yes'][['Provider', 'State', 'Confidence']]

    return health_report, flagged_list

iface = gr.Interface(
    fn=process_providers,
    inputs=gr.File(label="Upload Raw Provider CSV"),
    outputs=[
        gr.Dataframe(label="Provider Health Report"),
        gr.Dataframe(label="Flagged Providers List")
    ],
    title="Aegis Mind",
    description="Aegis Mind: Upload raw provider data, calculate confidence scores, and generate health & flagged provider reports."
)

iface.launch(share=True)
```
